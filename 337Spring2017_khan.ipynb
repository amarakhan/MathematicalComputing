## Introduction
Primitive Pythagorean Triples are positive integer solutions a,b,c such that $a^2$ + $b^2$ = $c^2$. Also, the greatest common divisor of a and b needs to equal 1. A set of pythagorean triples must have values that are positive integers, because there is no negative length. According to Euclid, there are infinite many pythagorean triples. There are infinite many primitive pythagorean triples, because there are infinite many pairs of m and n where m and n are relatively prime. 

We have created a function mygcd, which returns the greatest common denominator of a and b. We have created another function is_square, which returns True if the input is a square number. By using these two functions, we can test values for a in b in a specified range, and see which values are pythagorean triples. We have created several plots to visualize patterns for the triples found.

### mygcd
In this function we are subtracting the larger number by the smaller number until they are equal to each other. The function returns *a* as the greatest common divisor.

```python
from math import *
import matplotlib.pyplot as plt

def mygcd(a, b):
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    return a
```

### is_square

We took in an x to determine if it is a perfect square. We first took the square root, *sqrt*, of the x and then took the *floor* of our root.  By taking the *floor* of the root we eliminated any decimal points it may have had.  After we had our root we squared our result to compare it to the original x.  We returned *True* if they equaled each other, meaning it was a perfect square, and *False* otherwise.


```python
def is_square(x):
    square = floor(sqrt(x))
    if square**2 == x:
        return True
    else:
        return False
```

### Main code

* We created 3 lists to keep track of all of our *a*, *b*, and *c* values.

* We looped through from 1-5000 to find values of primative pythagorean triples. We filtered our results by setting *b* equal to *a* so we don't get duplicates from the lower region.

* The first step we take is determaining the greatest common divisor of *a* and *b* by callin the *mygcd* function.  If mygcd doesn't return 1 we continue on with our search.  If it does return 1 we move on to the next step.

* We compute the value of *$c^2$* by adding up *$a^2$* + *$b^2$*.

* After we have computed *$c^2$* we use the *is_square* function to test if it's a perfect square.

* If it's a perfect square we append our lists with the respective values of *a*, *b*, and the *floor* *square root* of *c*.

* Once all the points are collected we plot our graphs.


```python
aList, bList, cList = [], [], []
al, bl, cl = [], [], []
for a in range(1, 5001):
    for b in range(1,5001):
        if (mygcd(a, b)) == 1:
            c = a**2 + b**2
            if is_square(c) == True:
                if b >= a:
                    al.append(a)
                    bl.append(b)
                    cl.append(floor(sqrt(c)))
                aList.append(a)
                bList.append(b)
                cList.append(floor(sqrt(c)))
```

** Project Background **

Mahcine epsilon is relevant when we talk about floating point and precision. Machine epsilon gives an upper bound on the rounding error from floating point arithmatic. A computer can only store so many bits from a number. When faced with computing $\frac{1}{3}+\frac{1}{3}+\frac{1}{3}$ for example, the computer sees $0.33\bar{3}+0.33\bar{3}+0.33\bar{3}$. We would compute that to $\frac{1}{3}+\frac{1}{3}+\frac{1}{3}=1$, but a computer would compute it as $0.99\bar{9}$ depending on how many bits the computer can store. There will be an error from floating point arithmetic. In many cases, the precision from the computer is enough. 

```python
from math import *
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Part A: Finding the Floating Point Parameters

In class, we defined a function called find_epsilon. This function returns two values, n-1 and machine epsilon. **n-1** equates to the mantissa. We found $\epsilon$ by using a while loop. With the condition, 1+x does not equal 1, we increment 1 and reassign x to the value of $2^{-n}$. Machine epsilon is the difference between 1 and the next largest number that can be stored in that way.

** Find Epsilon**

```python
def find_epsilon():
    n = 1
    x = 2**(-n)
    while((1+x) != 1):
        n += 1
        x = 2**(-n)
    print("Mantissa: {}".format(n-1))
    print("Epsilon: {}".format(2**(1-n)))
    return n-1 , 2**(1-n)
    
find_epsilon()
```

**Find Largest Floating Point Number**

To find the largest floating point number, we increase x until we reach overflow. We increment variable m, after every reassignment of x. In the except case, we print out the value of x before overflow is reached. This will end up being the largest floating point number.

```python
def find_largest():
    z,m = 2.0,1.0
    x = (z-2**-52)*2.0**m
    while(x != 0):
        try:
            x = (z-2**-52)*(2.0**m)
            m += 1
        except:
            x = (z-2**-52)*(2.0**(m-1))
            print('Exponent: {}'.format(m))
            print('Largest Floating Point Number: {}'.format(x))
            break
```

```python
find_largest()
```

It is interesting to note that numbers greater than $2^{1023}(2-2^{-52})$ result in overflow. In other words, the magnitude is too great to represent. 

**Find Smallest Floating Point Number**

To find the smallest floating point number, we want to get as close to 1.0, without resulting in underflow. This is why we have set the exponent of 2 to 1-m-b. By incrementing b by 1 while checking to make sure the number x is not yet 0, we find the smallest floating point number without reaching underflow. 

```python
def find_smallest():
    m,b = 52, 1
    x = 2**(1-m-b)
    while (x != 0):
        b += 1
        x = 2**(1-m-b)
        exp = 1-m-(b-1)
    print("Smallest Floating Point Number: {},".format(2**(1-m-(b-1))), "It can also be written as 2^{}".format(exp))
    return 2**(1-m-(b-1))
```

```python
find_smallest()
```

It is interesting to note that floating point operations that result in a magnitude less than $2^{-1023}(1+0)$ will cause underflow. The number was too small to represent in other words.

## Exercise 1. Displaying the Transient Behavior

For the function mayfly, I created an empty numpy array to store the population values at various times t. The function has three parameters, x, b, and t. When the function is called, different values can be chosen for time t, the growth parameter b, and an initial value x for the population function. A for loop is used to index through t by i. The population at i(a number in the range of time) is set equal to x. x is set equal to b*x*(1-x). For each value of x at i, population is plotted on the graph. This is a simple way to represent the mayfly population. This mayfly model will be run for various b values to determine it's influence. on the function for mayfly population growth.

```python
def mayfly(x, b, t):
    population = np.empty(t)
    for i in range(t):
        population[i] = x
        x = b*x*(1-x)
    plt.figure(figsize = (10, 6))
    plt.title("population vs. time with b = {}".format(b))
    plt.xlabel("time t")
    plt.ylabel("population")
    plt.margins(.01)
    plt.plot(population, '.')
```

## Exercise 3. Separation of Trajectories

Function mayfly_sep, calculates and plots and separation of trajectories over time. The function is based on three parameters, x, b, and t. The separation is initially set as $10^{-5}$. m is assigned to x, n is assigned to x+separation, and s is assitned to an array of [separation]. I use a for loop to index through range t by i. Within the for loop, I assign n to b*n*(1-n) and m to b*m*(1-m). These are for population functions. n is calculated without separation, and m is calculated with the separation. I take the absolute value of the m-n. This will give me the separation of the populations m and n. I append the array s. The array gets plotted with the semilogy function imported from math. 

```python
def mayfly_sep(x,b,t):
    separation = 10**(-5)
    m,n,s = x, x+separation, [separation]
    for i in range(t):
        n = b*n*(1-n)
        m = b*m*(1-m)
        dist = abs(m-n)
        s.append(dist)
        plt.semilogy(s,'g.', ms = 4)
        title("separation of trajectories vs. time for b = {} and t = {}".format(b,t))
        plt.xlabel("time t")
        plt.ylabel("separation of trajectories")
```

For report 4, we looked at Newton's method in the complex plane. Newton's method is a method using the first few terms of the Taylor Series to approximate the root of a function. We applied Newton's method in exercise one when we were looking at the roots of complex numbers.

## Exercise 1

In exercise one we created a numpy array named roots in class. Roots is comprised of three complex functions. We created two linspace arrays(x and y). x ranges from -1 to 1 and contains 500 equally spaced values within the range. y ranges from -1j to 1j and contains 500 equally spaced values within the range. We assign X and Y to a numpy meshgrid of x and y (the linspace arrays). Meshgrid, in python, creates a grid out of x and y values. niters is set to 200, because we will be going up to 200 iterations when applying newtons method. We then include a for loop, iterating through the range of niters=200. For each i in the range, we apply Newton's method, setting Z equal to $Z-(Z^3 -1)/(3Z^2)$. Convergance is checked for by setting tol equal to 0.01, then assigning 3 numpy arrays to the absolute value of Z-roots[] < tol. These 3 numpy arrays, root1, root2, and root3 will be plotted in Exercise 2. 

```python
roots = np.array([complex(np.cos(2*np.pi*i/3), np.sin(2*np.pi*i/3)) for i in range(3)])
print(roots)
npts = 500
x = np.linspace(-1,1,npts)
y = np.linspace(-1j,1j,npts)
X,Y = np.meshgrid(x,y)
Z = X+Y
niters = 200
for i in range(niters):
    Z = Z- (Z**3 -1)/(3*Z**2) #where we are applying newton's method
tol = 0.01 #checking for convergence below
root1 = np.abs(Z-roots[0]) < tol
root2 = np.abs(Z-roots[1])<tol
root3 = np.abs(Z-roots[2])<tol
```

Also, we add 1 to the z0, z1, and z2 arrays at the indices from root0, root1, and root2. the itercount gets incremented by1. I checked that niters will be the largest possible number for iterations taken to converge. Thus I divided z0, z1, and z2 all by niters to get the arrays contents to be from 0 to 1.
```python
roots = np.array([complex(np.cos(2*np.pi*i/3), np.sin(2*np.pi*i/3)) for i in range(3)])
npts = 500
x = np.linspace(-1,1,npts)
y = np.linspace(-1j,1j,npts)
X,Y = np.meshgrid(x,y)
Z = X+Y
niters = 60
itercount = 0
tol = 0.01 
z0 = np.zeros((len(x),len(y))) #color this [1,1,0]
z1 = np.zeros((len(x),len(y))) #color this [0,1,1]
z2 = np.zeros((len(x),len(y))) #color this [1,0,1]
while itercount < niters :
    Z = Z- (Z**3 -1)/(3*Z**2) #where we are applying newton's method
    root0 = np.abs(Z-roots[0])<tol
    root1 = np.abs(Z-roots[1])<tol
    root2 = np.abs(Z-roots[2])<tol
    z0[root0] += 1
    z1[root1] += 1
    z2[root2] += 1
    itercount +=1

#print(z0[100])  
#prints out 0 if root did not converge, but if it did converge..prints out iterations 
#it has been since root converged.. so technically it is niters-#inarray..
#print(z0[100].size)
#print(np.amax(z0))
#print(np.amax(z1))
#print(np.amax(z2))
#checking for greatest number of iterations it converged to in 2D array so I know what to devide by
#dividing 

z0/=niters
z1/=niters
z2/=niters

img = np.zeros((npts,npts, 3))
img[:, :, 1] = z2+z0
img[:, :, 2] = z2
img[:,:,0] = z1+z0


#img[z1] = [0,1,1]
#img[z2] = [1,0,1]
plt.figure(figsize=(8,8))
plt.imshow(img);
```

# Part A: One Star System

In dP_dt, we had to account for the gravitational constant (g), the mass of the planet (m), the mass of one star at the center of the system (M), and the distance betweem the planet and star (r). dP_dt yields the DE we need to implement later. In dP_dt, we found the acceleration of the planet and returned the acceleration and velocity. 

\begin{equation}
     \frac{d\vec{P}}{dt}=\begin{bmatrix}
         V_{x} \\
         V_{y} \\
         \frac{-GMx}{{\sqrt{x^2 + y^2}}^3}\\
         \frac{-GMy}{{\sqrt{x^2 + y^2}}^3}
        \end{bmatrix}
  \end{equation}

```python
def dP_dt(P, t):
    G = 1.0
    M = 1.0
    x, y, vx, vy = P
    r3 = np.hypot(x, y)**3
    ax, ay = -G*M*x/r3, -G*M*y/r3
    return np.array([vx, vy, ax, ay])
```

### Heun's Method
The global error is O(h2)

```python
def heun_path(tmax, h, P0):
    steps = int(tmax/h)
    path = np.empty((steps+1, 4))
    P = np.copy(P0)
    path[0] = P
    for i in range(steps):
        F1 = h*dP_dt(P, i*h)
        F2 = h*dP_dt(P + F1, (i+1)*h)
        P += (F1 + F2)/2.
        path[i+1] = P
    return path

def plot_heun(tmax, h, P):
    path = heun_path(tmax, h, P)
    plt.plot(path[:,0], path[:,1])
    plt.axis('equal')
    plt.plot(0, 0, 'yo', ms=10)
```

```python
P0 = np.array([1., 0., 0., 1.])
print(heun_path(1, .1, P0))
```

```python
plt.figure(figsize=(6,6))
P0 = np.array([1.,0.,0.,1.])
plot_heun(10000, 0.1, P0)
plt.title("Heun's Method for One Star System")
plt.grid()
```

Heun's method has a larger global error than the Runge-Kutta 4th order method. Although, it is still more accurate than Euler's method for differential equations.

# Part B: Two Star System

The two star system is a planet orbiting 2 stars, also known as a circumbinary planet. This is caused by the short orbit of some stars. The only way a planet can form is to form outside the orbit of the two stars. On NASA's Kepler mission, they discovered this system 200 light years from earth. The two star system has also been protrayed in Star Wars on Tatooine. dP_dt had to be modified to account for the change in the distance, change in acceleration, and the two masses of the stars. heun_path had to be slightly modified into path_2star. 

```python
def dPdt_2star(P,t):
    G = 1.0
    M1 = 0.5
    M2 = 0.5
    r = 0.5
    x, y, vx, vy = P
    x1, y1, x2, y2 = r*cos(t), r*sin(t), -r*cos(t), -r*sin(t)
    r3a, r3b = (math.sqrt((x-x1)**2 + (y-y1)**2))**3, (math.sqrt((x-x2)**2 + (y-y2)**2))**3
    ax, ay = -G*M1*(x-x1)/r3a - G*M2*(x-x2)/r3b, -G*M1*(y-y1)/r3a - G*M2*(y-y2)/r3b
    return np.array([vx, vy, ax, ay])
```

```python
def path_2star(tmax, h, P0):
    steps = int(tmax/h)
    path = np.empty((steps, 4))
    P = np.copy(P0)
    for i in range(steps):
        t=i*h
        path[i] = P
        F2 = (h*dPdt_2star(P,t))+  P
        P += h*(dPdt_2star(P,t) + dPdt_2star(F2,t+h))/2
    return path
print(path)
```

```python
def rng(x, npts):
    a = 427419669081
    m = 999999999989
    u = np.empty(npts)
    u[0] = x/m
    for i in range(1, npts):
        x = (a*x) % m
        u[i] = x/m
    return u
```

```python
wordcount = {}  #how to declare empty dictionary

for word in words:
    if word in wordcount:
        wordcount[word] += 1
    else:
        wordcount[word] = 1
```

```python
def get_count(x):
    return x[1]
```

```python
sorted_words = sorted(wordcount.items(), key=get_count, reverse=True)
```
